```python
# Dependencies
import requests
import json
import pandas as pd
from yelpapi import YelpAPI
import matplotlib.pyplot as plt

# Google API Key
from config import ykey_access_token
yelp_api = YelpAPI(ykey_access_token)
 
```

```python
store_locations = pd.read_csv("Resources/projectbeggarsstart_geo.csv")
store_locations
```

```python
target_categories = "pizza"
target_radius = 500
target_limit = 50

ID = []
name = []
rating = []
review_count = []
lat = []
lng = []
city = []
state = []
zipcode = []
transactions = []
distance = []
category = []

for index, row in store_locations.iterrows():

    target_id = row["SQL_ID"]

    # Search Yelp
    search_results = yelp_api.search_query(latitude=row["Latitude"],longitude=row["Longitude"],categories=target_categories,radius=target_radius,limit=target_limit)

    for result in search_results["businesses"]:
        if "Beggar" in result["name"]:
            ID.append(target_id)
            rating.append(result["rating"])
            review_count.append(result["review_count"])
            name.append(result["name"])
            lat.append(result["coordinates"]["latitude"])
            lng.append(result["coordinates"]["longitude"])
            city.append(result["location"]["city"])
            state.append(result["location"]["state"])
            zipcode.append(result["location"]["zip_code"])
            transactions.append(result["transactions"])
            distance.append(result["distance"])
            category.append(result["categories"][0]["title"])
```

```python
beggars_reviews = pd.DataFrame({
    "ID" : ID,
    "Name" : name,
    "Category" : category,
    "Review Count" : review_count,
    "Rating" : rating,
    "Transactions" : transactions,
    "Distance" : distance,
    "Latitude" : lat,
    "Longitude" : lng,
    "City": city,
    "State": state,
    "ZipCode": zipcode
})
beggars_reviews.to_csv("Resources/beggarsreviews.csv", index=False)

beggars_reviews
```
