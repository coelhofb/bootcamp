```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_borrower = "data/Borrower.txt"
file_loan = "data/Loan.txt"

file_borrower_prod = "data/Borrower_Prod.txt"
file_loan_prod = "data/Loan_Prod.txt"


# Read files and store into pandas dataframes
borrower_df = pd.read_csv(file_borrower,sep='\t',header=(0))
loan_df = pd.read_csv(file_loan,sep='\t',header=(0))

borrower_prod_df = pd.read_csv(file_borrower_prod,sep='\t',header=(0))
loan_prod_df = pd.read_csv(file_loan_prod,sep='\t',header=(0))
```

```python
bins_loanAmount = [-1, 14953, 18951, 20852, 22122, 24709, 28004, 1000000]
labels_loanAmount = ["0 to 14953", "14953 to 18951", "18951 to 20852", "20852 to 22122", "22122 to 24709", "24709 to 28004", "> 28004"]
bins_interestRate = [-1, 7.17, 10.84, 12.86, 14.47, 15.75, 18.05, 1000000]
labels_interestRate = ["0 to 7.17", "7.17 to 10.84", "10.84 to 12.86", "12.86 to 14.47", "14.47 to 15.75", "15.75 to 18.05", "> 18.05"]
bins_monthlyPayment = [-1, 382, 429, 495, 529, 580, 649, 708, 847, 1000000]
labels_monthlyPayment = ["0 to 382", "382 to 429", "429 to 495", "495 to 529", "529 to 580", "580 to 649", "649 to 708", "708 to 847", "> 847"]
bins_annualIncome = [-1, 49402, 50823, 52089, 52885, 53521, 54881, 55520, 57490, 1000000]
labels_annualIncome = ["0 to 49402", "49402 to 50823", "50823 to 52089", "52089 to 52885", "52885 to 53521", "53521 to 54881", "54881 to 55520", "55520 to 57490", "> 57490"]
bins_dtiRatio = [-1, 9.01, 13.42, 15.92, 18.50, 21.49, 22.82, 24.67,1000000]
labels_dtiRatio = ["0 to 9.01", "9.01 to 13.42", "13.42 to 15.92", "15.92 to 18.50", "18.50 to 21.49", "21.49 to 22.82", "22.82 to 24.67", "> 24.67"]
bins_lengthCreditHistory = [-1, 8, 1000000]
labels_lengthCreditHistory = ["0 to 8", "> 8"]
bins_numTotalCreditLines = [-1, 1, 2, 1000000]
labels_numTotalCreditLines = ["0 to 1", "2", "> 2"]
bins_numOpenCreditLines = [-1, 3, 5, 1000000]
labels_numOpenCreditLines = ["0 to 3", "3 to 5", "> 5"]
bins_numOpenCreditLines1Year = [-1, 3, 4, 5, 6, 7, 9, 1000000]
labels_numOpenCreditLines1Year = ["0 to 3", "4", "5", "6", "7", "7 to 9", "> 9"]
bins_revolvingBalance = [-1, 11912, 12645, 13799, 14345, 14785, 15360, 15883, 16361, 17374, 18877, 1000000]
labels_revolvingBalance = ["0 to 11912", "11912 to 12645", "12645 to 13799", "13799 to 14345", "14345 to 14785", "14785 to 15360", "15360 to 15883", "15883 to 16361", "16361 to 17374", "17374 to 18877", "> 18877"]
bins_revolvingUtilizationRate = [-1, 49.88, 60.01, 74.25, 81.96, 1000000]
labels_revolvingUtilizationRate = ["0 to 49.88", "49.88 to 60.01", "60.01 to 74.25", "74.25 to 81.96", "> 81.96"]
bins_numDerogatoryRec = [-1, 0, 1, 1000000]
labels_numDerogatoryRec = ["0", "1", "> 1"]
bins_numDelinquency2Years = [-1, 0, 1000000]
labels_numDelinquency2Years = ["0", "> 0"]
bins_numChargeoff1year = [-1, 0, 1000000]
labels_numChargeoff1year = ["0", "> 0"]
bins_numInquiries6Mon = [-1, 0, 100000]
labels_numInquiries6Mon = ["0", "> 0"]

cleaned_df["loanAmount_bins"] = pd.cut(cleaned_df["loanAmount"], bins_loanAmount, labels=labels_loanAmount)
cleaned_df["interestRate_bins"] = pd.cut(cleaned_df["interestRate"], bins_interestRate, labels=labels_interestRate)
cleaned_df["monthlyPayment_bins"] = pd.cut(cleaned_df["monthlyPayment"], bins_monthlyPayment, labels=labels_monthlyPayment)
cleaned_df["annualIncome_bins"] = pd.cut(cleaned_df["annualIncome"], bins_annualIncome, labels=labels_annualIncome)
cleaned_df["dtiRatio_bins"] = pd.cut(cleaned_df["dtiRatio"], bins_dtiRatio, labels=labels_dtiRatio)
cleaned_df["lengthCreditHistory_bins"] = pd.cut(cleaned_df["lengthCreditHistory"], bins_lengthCreditHistory, labels=labels_lengthCreditHistory)
cleaned_df["numTotalCreditLines_bins"] = pd.cut(cleaned_df["numTotalCreditLines"], bins_numTotalCreditLines, labels=labels_numTotalCreditLines)
cleaned_df["numOpenCreditLines_bins"] = pd.cut(cleaned_df["numOpenCreditLines"], bins_numOpenCreditLines, labels=labels_numOpenCreditLines)
cleaned_df["numOpenCreditLines1Year_bins"] = pd.cut(cleaned_df["numOpenCreditLines1Year"], bins_numOpenCreditLines1Year, labels=labels_numOpenCreditLines1Year)
cleaned_df["revolvingBalance_bins"] = pd.cut(cleaned_df["revolvingBalance"], bins_revolvingBalance, labels=labels_revolvingBalance)
cleaned_df["revolvingUtilizationRate_bins"] = pd.cut(cleaned_df["revolvingUtilizationRate"], bins_revolvingUtilizationRate, labels=labels_revolvingUtilizationRate)
cleaned_df["numDerogatoryRec_bins"] = pd.cut(cleaned_df["numDerogatoryRec"], bins_numDerogatoryRec, labels=labels_numDerogatoryRec)
cleaned_df["numDelinquency2Years_bins"] = pd.cut(cleaned_df["numDelinquency2Years"], bins_numDelinquency2Years, labels=labels_numDelinquency2Years)
cleaned_df["numChargeoff1year_bins"] = pd.cut(cleaned_df["numChargeoff1year"], bins_numChargeoff1year, labels=labels_numChargeoff1year)
cleaned_df["numInquiries6Mon_bins"] = pd.cut(cleaned_df["numInquiries6Mon"], bins_numInquiries6Mon, labels=labels_numInquiries6Mon)

# cleaned_df.loc[:, ["numInquiries6Mon", "numInquiries6Mon_bins"]]
```
